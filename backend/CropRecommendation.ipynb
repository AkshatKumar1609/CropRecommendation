```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
crop = pd.read_csv('Crop_recommendation.csv')
```

```python
crop.head()
```

```python
crop.info()
```

```python
crop.describe()
```

```python
X = crop.drop('label', axis=1)
y = crop['label']
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
reverse_mapping = dict(zip(le.transform(le.classes_), le.classes_))
print("Reverse mapping (label → category):", reverse_mapping)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
trf1 = ColumnTransformer([("standardscaler",StandardScaler(), ["N","P","K","temperature","humidity","ph","rainfall"])],remainder="passthrough")
trf2 = RandomForestClassifier()
```

```python
pipe = Pipeline([
    ("trf1", trf1),
    ("trf2", trf2)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import pickle
pickle.dump(pipe,open('CropRecommendation','wb'))
```
